```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cdf = pd.read_csv('classical_csv.csv')
qdf = pd.read_csv('quantum_major_csv.csv')
```

```python
columns = ["watchers", "forks", "issues", "commits"]
cvalues = []
qvalues = []
for column in columns:
#     print(column)
    print(f"{column}: classical {cdf[column].sum()}, quantum {qdf[column].sum()}" )
#     print(f"Avg. {column}: classical {cdf[column].sum()/61}, quantum {qdf[column].sum()/61}" )
    cvalues.append(cdf[column].sum()*(10000)/57)
    qvalues.append(qdf[column].sum()*(10000)/62)
print(cvalues)
print(qvalues)
```

```python
plt.bar(np.arange(4), cvalues, width = 0.4, label = "classical")
plt.bar(np.arange(4) + 0.4, qvalues, width = 0.4, label = "quantum")

# plt.ylim(bottom=0, top= 50000)

plt.xticks(np.arange(4), columns)
plt.xlabel("Repository Statistics")
plt.ylabel("Average number")
plt.title("Comparision of Major classical and quantum repositories")
plt.legend()
plt.show()
```

```python
import ast
# print(cdf["unique labels"][1].strip("['']").split(","))

clabels = []
common_labels = []
for i in range(57):
    label_list = ast.literal_eval(cdf["unique labels"][i])
    for item in label_list:
        clabels.append(item)
        common_labels.append(item)
        
print(set(clabels))
```

```python
qlabels = []
for i in range(62):
    label_list = ast.literal_eval(qdf["unique labels"][i])
    for item in label_list:
        qlabels.append(item)
        common_labels.append(item)
        
print(set(qlabels))
```

```python
list(set(clabels).intersection(qlabels))
```
